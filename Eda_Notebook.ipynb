```python
num=reading_habits_1.select_dtypes(include=["int64"])
```

```python
#Graficamos las numericas 
num.hist(figsize=(5,5)); 
```

```python
#Analizamos las variables numericas, para ver cuales son normales y cuales no normales  
num_norm=pd.DataFrame()
num_no_norm=pd.DataFrame()
for i in num:
    if -2<num[i].skew()<2  and -3<num[i].kurt()<3:
        num_norm[i]=num[i]
    else:
        num_no_norm[i]=num[i]
    

print(f"las normales son: {num_norm.columns}")
print(f"las no normales son: {num_no_norm.columns}")
```

```python
def norm_desc(i):
    return {"name":i.name,"mean":i.mean(),"std":i.std(),"skew":i.skew(),"kurtosis":i.kurtosis()}
```

```python
norm_desc(num_norm.Age)
```

```python
fig, axe = plt.subplots(2) # aqui vemos un ejemplo de mezclar seaborn y Mtplot. Axe es la nexo
sns.boxplot(num_norm.Age,ax=axe[0],orient="hor")
sns.histplot(num_norm.Age,ax=axe[1])
axe[1].axvline(num_norm.Age.mean(), color='r', linestyle='--', label="Mean")
axe[1].axvline(num_norm.Age.median(), color='g', linestyle='-', label="Median")
axe[1].axvline(num_norm.Age.mode().values[0], color='b', linestyle='-', label="Mode")
axe[1].legend();
```
